```python
# 📌 Import Library dan Load Dataset 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns

sns.set_theme(style="whitegrid")

df = sns.load_dataset('tips')

print("📊 Shape Dataset :", df.shape)
print("\n📋 Tipe Data")
print(df.dtypes)

print("\n📈 Statistik Deskriptif")
print(df.describe().round(2)) 
```

```python
# 📌 Membuat Dashboard Visualisasi

fig = plt.figure(figsize=(16,12))
fig.suptitle(
    "Dashboard Analisis Dataset Tips",
    fontsize=18,
    fontweight='bold'
)
gs = gridspec.GridSpec(
    2,2,
    figure=fig,
    hspace=0.35,
    wspace=0.3
)
ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
ax3 = fig.add_subplot(gs[1,0])
ax4 = fig.add_subplot(gs[1,1])

# =========================
# Grafik 1 - Bar Chart
# =========================
avg_by_day.plot(
    kind='bar',
    color='#D8B4FE',
    edgecolor='white',
    ax=ax1
)
ax1.set_title('Rata-rata Total Bill per Hari')
ax1.set_xlabel('Hari')
ax1.set_ylabel('Total Bill (USD)')

# =========================
# Grafik 2 - Histogram
# =========================

sns.histplot(
    data=df,
    x='total_bill',
    kde=True,
    bins=20,
    color='#D8B4FE',
    ax=ax2
)
ax2.axvline(
    df['total_bill'].mean(),
    color='#F9A8D4',
    linestyle='--',
    linewidth=2,
    label='Mean'
)
ax2.axvline(
    df['total_bill'].median(),
    color='#BAE6FD',
    linestyle='--',
    linewidth=2,
    label='Median'
)
ax2.legend()
ax2.set_title('Distribusi Total Bill')

# =========================
# Grafik 3 - Boxplot
# =========================

sns.boxplot(
    data=df,
    x='day',
    y='total_bill',
    hue='sex',
    palette=['#D8B4FE','#F9A8D4'],
    ax=ax3
)
        
ax3.set_title('Distribusi Tagihan per Hari dan Gender')

# =========================
# Grafik 4 - Scatter Plot
# =========================

sns.scatterplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='time',
    size='size',
    sizes=(40,200),
    palette=['#D8B4FE','#F9A8D4'],
    alpha=0.75,
    ax=ax4
)

sns.regplot(
    data=df,
    x='total_bill',
    y='tip',
    scatter=False,
    color='#BAE6FD',
    line_kws={'linewidth':3},
    ax=ax4
)

ax4.set_title('Hubungan Total Bill dan Tip')

plt.show() 
```

## 📌 Grafik 2 - Histogram

### 🔎 Analisis (What? So What? Now What?)

**What?** Histogram menunjukkan distribusi total tagihan pelanggan yang sebagian besar berada pada rentang nilai rendah hingga menengah. Distribusi terlihat miring ke kanan dengan beberapa nilai tagihan yang cukup tinggi.

**So What?** Kondisi ini menunjukkan bahwa sebagian besar pelanggan melakukan transaksi dalam jumlah yang relatif moderat, sedangkan transaksi bernilai besar hanya terjadi pada sebagian kecil pelanggan.

**Now What?** Perlu dilakukan analisis lebih lanjut terhadap pelanggan dengan transaksi tinggi untuk mengetahui karakteristik atau pola pembelian mereka.
