## 🚢 Studi Kasus Dataset Titanic

Dataset Titanic merupakan salah satu dataset yang sering digunakan dalam pembelajaran Data Science. Dataset ini berisi informasi penumpang kapal Titanic seperti usia, jenis kelamin, kelas penumpang, dan status keselamatan.

Pada praktikum ini dilakukan beberapa analisis sederhana untuk memahami karakteristik data menggunakan Pandas.

```python
# Import Library
import pandas as pd
import numpy as np

print("=" * 60)
print("📊 ANALISIS DATASET TITANIC")
print("=" * 60)

# Membaca Dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("\n✅ Dataset Titanic berhasil dimuat.")
print(f"Jumlah Data : {df.shape[0]} baris")
print(f"Jumlah Kolom : {df.shape[1]} kolom")

# ==================================================
# Q1. Jumlah Baris dan Kolom Dataset
# ==================================================

print("\n" + "=" * 60)
print("🔹 Q1. Jumlah Baris dan Kolom Dataset")
print("=" * 60)

baris, kolom = df.shape

print(f"Dataset Titanic memiliki {baris} baris dan {kolom} kolom.")

# ==================================================
# Q2. Missing Value
# ==================================================

print("\n" + "=" * 60)
print("🔹 Q2. Missing Value pada Dataset")
print("=" * 60)

missing = df.isnull().sum()

print(missing)

print("\n📌 Terdapat beberapa kolom yang masih memiliki data kosong,")
print("terutama pada kolom Age, Cabin, dan Embarked.")

# ==================================================
# Q3. Persentase Penumpang Selamat
# ==================================================

print("\n" + "=" * 60)
print("🔹 Q3. Persentase Penumpang yang Selamat")
print("=" * 60)

persentase_selamat = df["Survived"].mean() * 100

print(f"Persentase penumpang yang selamat adalah {persentase_selamat:.2f}%")

# ==================================================
# Q4. Rata-rata Usia Wanita Kelas 1 yang Selamat
# ==================================================

print("\n" + "=" * 60)
print("🔹 Q4. Rata-rata Usia Wanita Kelas 1 yang Selamat")
print("=" * 60)

rata_usia = df[
    (df["Sex"] == "female") &
    (df["Pclass"] == 1) &
    (df["Survived"] == 1)
]["Age"].mean()

print(f"Rata-rata usia penumpang wanita kelas 1 yang selamat adalah {rata_usia:.2f} tahun.")

# ==================================================
# Q5. Kelas dengan Tingkat Keselamatan Tertinggi
# ==================================================

print("\n" + "=" * 60)
print("🔹 Q5. Kelas dengan Tingkat Keselamatan Tertinggi")
print("=" * 60)

keselamatan = df.groupby("Pclass")["Survived"].mean()

print(keselamatan)

kelas_tertinggi = keselamatan.idxmax()

print(f"\nKelas dengan tingkat keselamatan tertinggi adalah Kelas {kelas_tertinggi}.")

# ==================================================
# Q6. Hasil GroupBy Berdasarkan Kelas Penumpang
# ==================================================

print("\n" + "=" * 60)
print("🔹 Q6. Hasil GroupBy Berdasarkan Kelas Penumpang")
print("=" * 60)

print(df.groupby("Pclass")["Survived"].mean())

print("\n✅ Analisis Dataset Titanic selesai dilakukan.")  
```
